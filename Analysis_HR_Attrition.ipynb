```python
#importing all python libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\DELL\Downloads\ALLDATASET\HR_Analytics.csv',encoding= 'unicode_escape')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df.head()
```

```python
ax = sns.countplot(x = 'Attrition',data = df)
plt.title("Count of Attrition")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x='Attrition',data=df,hue='Gender')
plt.title("Count of Attrition by Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python

plt.figure(figsize=(5,5))
sns.countplot(x='Attrition',data=df,hue='AgeGroup')
plt.title("Count of Attrition by AgeGroup")
plt.show()
```

```python
# USE OF PIE PLOT
x=df['EducationField'].value_counts().index
y=df['EducationField'].value_counts().values

my_explode=[0.2,0,0,0,0,0]
plt.figure(figsize=(5,5))
plt.pie(y,labels=x,autopct="%0.2f%%",explode=my_explode,shadow=True)
plt.title("Count of various EducationalField")
plt.legend()
plt.show()
```

```python
#USE OF BAR PLOT
plt.figure(figsize=(5,5))
sns.countplot(x='Attrition',hue='Gender',data=df)
plt.title("Count of Attrition by Gender")
plt.savefig("Count of Attrition by Gender.jpg") #The Graph is Saved which can be ued anywhere..
plt.show()
```

```python
# USE OF SCATTER PLOT
plt.figure(figsize=(5,5))
df.plot(kind='scatter',x='AgeGroup',y='YearsAtCompany')
plt.title("Count of YearsAtCompany by AgeGroup")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='SalarySlab',hue='Attrition',data=df)
plt.title("SalarySlab by Attrition")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df.isnull())
```
